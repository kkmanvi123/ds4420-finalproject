```python
import mlp_model as mlp
import pandas as pd
import numpy as np
```

## Collect Data

```python
data_dict = mlp.pull_data("data")

for name, df in data_dict.items():
        
    X = mlp.clean_feaures(df)

    print(f"\n{name.upper()}")

    # shapes
    print(f"Original shape: {df.shape}")
    print(f"Cleaned shape: {X.shape}")

    # print head
    # print("\nCleaned Head:")
    # display(X.head())

    # update dict
    data_dict[name] = X
```

## Perform PCA

```python
print("MRI PCA")
mri_pca = mlp.run_pca(df=data_dict["mri"], n=50)

print("\nFDG PCA")
fdg_pca = mlp.run_pca(df=data_dict["fdg_pet"], n=0.95)

print("\nAMYLOID PCA")
amyloid_pca = mlp.run_pca(df=data_dict["amyloid_pet"], n=0.95)

print("\nTAU PCA")
tau_pca = mlp.run_pca(df=data_dict["tau_pet"], n=0.95)
```
